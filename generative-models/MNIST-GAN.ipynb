```python
import torch as t
import torchvision
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from tqdm import tqdm
import matplotlib.pyplot as plt
import numpy as np
```

```python
BATCH_SIZE = 64
LEARNING_RATE = 1e-4
EPOCHS = 300
LATENT_SIZE = 64
HIDDDEN_SIZE = 256
```

```python
D = nn.Sequential(
        nn.Linear(28*28, HIDDDEN_SIZE),
        nn.LeakyReLU(),
        nn.Dropout(),
        nn.Linear(HIDDDEN_SIZE, HIDDDEN_SIZE),
        nn.LeakyReLU(),
        nn.Dropout(),
        nn.Linear(HIDDDEN_SIZE, 1),
        nn.Sigmoid()
    )

G = nn.Sequential(
        nn.Linear(LATENT_SIZE, HIDDDEN_SIZE),
        nn.ReLU(),
        nn.Linear(HIDDDEN_SIZE, HIDDDEN_SIZE),
        nn.ReLU(),
        nn.Linear(HIDDDEN_SIZE, 28*28),
        nn.Tanh()
    )
```

```python
mnist_ds = datasets.MNIST(
        root='data/',
        train=True,
        download=True,
        transform=transforms.Compose([
            transforms.ToTensor(),
            transforms.Normalize((0.5,), (0.5,))
        ])
    )

real_loader = DataLoader(mnist_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=4)
```

```python
def imshow(img):
    img = img / 2 + 0.5     # unnormalize - corresponds to transformation above.
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
```

```python
device = "cuda:0" if t.cuda.is_available() else "cpu"

D, G = D.to(device), G.to(device)

d_optim = optim.Adam(D.parameters(), lr=LEARNING_RATE)
g_optim = optim.Adam(G.parameters(), lr=LEARNING_RATE)

criterion = nn.BCELoss()

print(f'Training on device: {device}')
```

```python
"""
Train!
"""

for epoch in range(EPOCHS):

    running_d_loss, running_g_loss = 0.0, 0.0
    print(f'Begining Epoch {epoch + 1}/{EPOCHS}...')
    
    for real_imgs, _ in tqdm(real_loader):
        # Load MNIST images and labels to device
        real_labels = t.ones(len(real_imgs), 1).to(device)
        real_imgs = real_imgs.to(device)
        
        # Load latent vectors and labels to device
        z = t.randn(len(real_imgs), LATENT_SIZE).to(device)
        fake_labels = t.zeros(len(real_imgs), 1).to(device)

        #####################################
        #       Update Discriminator        #
        #####################################

        # Get probability scores for real and fake data
        real_imgs = real_imgs.view(-1, 28*28)
        real_logits = D(real_imgs)

        fake_imgs = G(z)
        fake_logits = D(fake_imgs)

        d_real_loss = criterion(real_logits, real_labels)
        d_fake_loss = criterion(fake_logits, fake_labels)
        d_loss = d_real_loss + d_fake_loss

        # Backpropagation and update
        d_optim.zero_grad()
        d_loss.backward()
        d_optim.step()

        #####################################
        #       Update Generator            #
        #####################################

        # Load another batch of latent vectors device
        z = t.randn(len(real_imgs), LATENT_SIZE).to(device)

        # Get generated images and and record loss
        fake_imgs = G(z)
        fake_logits = D(fake_imgs)
        g_loss = criterion(fake_logits, real_labels)

        # Backpropagation and update
        g_optim.zero_grad()
        g_loss.backward()
        g_optim.step()

        # Keep track of losses and global step
        running_g_loss += g_loss.item()
        running_d_loss += d_loss.item()

    #####################################
    #       Log Info for Epoch          #
    #####################################

    log_str = f"\n{'Completed Epoch:':<20}{epoch + 1:<10}"
    # Value to normalize so we get loss/sample
    norm = len(mnist_ds)
    log_str += f"\n{'Discriminator Loss:':<20}{running_d_loss/norm:<10}"
    log_str += f"\n{'Generator Loss:':<20}{running_g_loss/norm:<10}\n"
    print(log_str)
    
    # Display G's images
    z = t.randn(36, LATENT_SIZE).to(device)
    generated_imgs = G(z).detach().cpu()

    img_grid = torchvision.utils.make_grid(generated_imgs.reshape(36, 1, 28, 28), nrow=6)
    imshow(img_grid)
    plt.show()
```
